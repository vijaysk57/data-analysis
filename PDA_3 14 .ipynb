seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mydata = { "Name" : ["Ram","Sam","Joe","Asha"],
          "Age" : [23,22,26,47],
          "salary" : [15000,20000,32000,49000],
          "Exp" : [2,7,3,10]
         }
```

```python
df = pd.DataFrame(mydata)
df.head()
```

histogram

```python
plt.figure(figsize = (6,5))
sns.histplot(df["salary"],kde = True, bins =2)
plt.title("Distribution of salary")
plt.show()
```

1. postitve skew,large salary value detected 
2. majarity is between 15000 to 30000
3. average salary s 20k
4. no otliners detected

correlation Matrix(heatmap):

```python
#step 1: filter numerical data
ndf = df.select_dtypes(include = ["number"])
ndf.head()
```

```python
plt.figure(figsize = (6,5))
sns.heatmap(ndf.corr(),cmap = "cividis",annot = True)
plt.title("correation matrix between sal ,age, exp")
plt.show()
```

1.it shows the correlation btw age sal and exp
2.there is a postivie correlation btw age and sal
3.the correlation btw age and exp is less

```python
plt.figure(figsize = (6,5))
sns.boxplot(x=df["Age"])
plt.title("Age Distribution")
plt.show()
```

1. abnormal oytlier is around 45
2. the average age is around 25

countplot

```python
mydata1 = { "Name" : ["Ram","Sam","Joe","Asha"],
          "Age" : [23,22,26,47],
          "salary" : [15000,20000,32000,49000],
          "Exp" : [2,7,3,10],
          "gender" :['M','F','M','F']
         }
df1 = pd.DataFrame(mydata1)
df1.head()
```

```python
plt.figure(figsize = (6,5))
sns.countplot(x=df1["Exp"],palette = "pastel", hue = df1["gender"])
plt.title("count of exp")
```

pair plot

```python
sns.pairplot(df1,hue='gender')
```
