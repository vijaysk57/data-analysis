ONE-HOT-ENCODING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv("C:\my files\Salary_EDA.csv")
df.head()
```

filter categorical features

```python
categorical_cols=["Education Level"]
```

define and apply encoder
1) encoder = non numerical data into numerical data

```python
encoder = OneHotEncoder(drop=None,sparse_output=False)
```

```python
encoded_data = encoder.fit_transform(df[categorical_cols])
```

```python
print(encoded_data)
```

the encoded data is in form of array.now we need to encoded data into a data frame with convert the encoded features into a dataframe with categoies as column name

```python
encoded_df = pd.DataFrame(encoded_data,columns=encoder.get_feature_names_out(categorical_cols))
encoded_df.head()
```

```python
encoded_df.drop(columns=["Education Level_nan"],inplace = True,axis = 1)
encoded_df.head()
```

```python
final_df = pd.concat([df,encoded_df],axis = 1)
final_df.head()
```
