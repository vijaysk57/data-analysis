```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hdf = pd.read_csv("C:\my files\hotel_bookings.csv")
```

```python
hdf.head()
```

```python
hdf.info()
```

```python
hdf.isnull().sum()
```

```python
hdf.dropna(inplace = True)
hdf.isnull().sum()
```

```python
hdf.describe(include = "all")
```

```python
plt.figure(figsize = (6,5))
sns.histplot(hdf["company"],kde = True, bins =10)
plt.title("Distribution of company")
plt.show()
```

```python
plt.figure(figsize = (6,5))
sns.histplot(hdf["arrival_date_month"],kde = True, bins =10)
plt.xticks(rotation=45)
plt.title("Distribution of hotel")
plt.show()
```

```python
plt.figure(figsize = (6,5))
sns.boxplot(x=hdf["lead_time"])
plt.title("lead time Distribution")
plt.show()
```

```python
hdf1= hdf.select_dtypes(include = ["number"])
hdf1.head()
```

```python
plt.figure(figsize = (6,5))
sns.heatmap(hdf1.corr(),cmap = "cividis",annot = True)
plt.title("correation matrix ")
plt.show()
```

```python
plt.figure(figsize = (6,5))
sns.countplot(x=hdf["arrival_date_day_of_month"],palette = "pastel")
plt.title("month data")
```

```python
sns.pairplot(hdf,hue="arrival_date_day_of_month")
plt.show()
```
